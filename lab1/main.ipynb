```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Housing.csv')
df.head()
```

```python
print(df.head())
print("Mean:\n",df.mean(numeric_only=True),"\n")
df.describe()
```

```python
print("Mode:\n",df.mode(numeric_only=True).iloc[0],"\n")
print("Median:\n",df.median(numeric_only=True),"\n")
print("Standard Deviation:\n",df.std(numeric_only=True),"\n")
```

```python
print(df.head())
range_values = df.max(numeric_only=True) - df.min(numeric_only=True)
print("Range:\n", range_values, "\n")

q1=df.quantile(0.25, numeric_only=True)
q3=df.quantile(0.75, numeric_only=True)
iqr=q3-q1
print("Q1 (25th percentile):\n", q1, "\n")
print("Q3 (75th percentile):\n", q3, "\n")
print("Interquartile Range (IQR):\n", iqr, "\n")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Example: Box plot for a specific column (e.g., 'price')
plt.figure(figsize=(10, 6))
sns.boxplot(data=df['price'])
plt.title('Box Plot for Price')
sns.boxplot(data=df.drop(columns=[ "price","area"]))
plt.title('Box Plot (excluding price and area)')
plt.show()
```

```python
## QQ plot for price
plt.figure(figsize=(6, 6))
stats.probplot(df["price"].dropna(), dist="norm", plot=plt)
plt.title("QQ Plot  price")
plt.tight_layout()
plt.show()
```

```python
# Histogram for price
plt.figure(figsize=(8, 4))
sns.histplot(df["price"], bins=30, kde=True)
plt.title("Histogram  price")
plt.xlabel("price")
plt.ylabel("count")
plt.tight_layout()
plt.show()
```
